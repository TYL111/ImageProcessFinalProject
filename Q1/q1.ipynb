```python
import cv2
import numpy as np
```

```python
coin = cv2.imread('01_image.jpg')
#轉灰階
coin_gray = cv2.cvtColor(coin,cv2.COLOR_BGR2GRAY)
#高斯模糊 為了減少誤判
coin_gray_guassion = cv2.GaussianBlur(coin_gray,(21,21),2)
#HoughCircle  param1
circles = cv2.HoughCircles(coin_gray_guassion,cv2.HOUGH_GRADIENT,dp=1.2,minDist=30,param1=50,param2=50,minRadius=100,maxRadius=300)
```

```python
coin_circled = coin.copy()
if circles is not None:
    circles = np.round(circles[0,:]).astype("int")
    for (x,y,r) in circles:
        cv2.circle(coin_circled,(x,y),r,(0,255,0),4)
cv2.imshow("",coin_circled)
cv2.waitKey(0)
cv2.destroyAllWindows
```
