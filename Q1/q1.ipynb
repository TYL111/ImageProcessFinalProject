```python
import cv2
import numpy as np
```

```python
coin = cv2.imread('01_image.jpg')
#轉灰階
coin_gray = cv2.cvtColor(coin,cv2.COLOR_BGR2GRAY)
#高斯模糊 為了減少誤判
coin_gray_guassion = cv2.GaussianBlur(coin_gray,(13,13),2)
#HoughCircle  
circles = cv2.HoughCircles(coin_gray_guassion,cv2.HOUGH_GRADIENT,dp=1,minDist=150,param1=80,param2=100,minRadius=100,maxRadius=300)
```

```python
coin_circled = coin.copy()
if circles is not None:
    circles = np.round(circles[0,:]).astype("int")
    #畫圓 
    for (x,y,r) in circles:
        cv2.circle(coin_circled,(x,y),r,(0,255,0),4)
cv2.imwrite('hough_circle_result.jpg', coin_circled)
```

```python
#Q1-1
print("Number of coins :", len(circles))
```

```python
coin_sizes = {"1":0, "10":0, "50":0}
for r in circles[:,2]:
    if r < 160:
        coin_sizes["1"] += 1
    elif r < 205:
        coin_sizes["10"] += 1
    else:
        coin_sizes["50"] += 1
total_value = coin_sizes["1"]*1 + coin_sizes["10"]*10 + coin_sizes["50"]*50

#Q1-2
print("總共: {} 元".format(total_value))
```
