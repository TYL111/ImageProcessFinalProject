```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def show_image(title, image):
    """
    
    :param title: 
    :param image: 
    """
    plt.figure()
    if len(image.shape) == 2:  # 
        plt.imshow(image, cmap='gray')
    else:  # 
        plt.imshow(image)
    plt.title(title)
    plt.axis('off')
    plt.show()

def motion_blur(image, kernel_size, angle):
    """
    
    :param image:  ()
    :param kernel_size: 
    :param angle: 
    :return: 
    """
    # 
    kernel = np.zeros((kernel_size, kernel_size)) # 
    center = kernel_size // 2 # 
    cv2.line(kernel, (center, 0), (center, kernel_size - 1), 1, thickness=1) # 
    show_image("Initial Kernel (Line)", kernel)

    # 
    rotation_matrix = cv2.getRotationMatrix2D((center, center), angle, 1) # 
    kernel = cv2.warpAffine(kernel, rotation_matrix, (kernel_size, kernel_size)) # 
    show_image("Rotated Kernel", kernel)

    # 
    kernel /= kernel.sum()

    # 
    blurred = cv2.filter2D(image, -1, kernel)
    show_image("Blurred Image", blurred)

    return blurred, kernel

def wiener_filter(image, kernel, K=0.01):
    """
    
    :param image:  ()
    :param kernel: 
    :param K: 
    :return: 
    """
    result = np.zeros_like(image)
    if len(image.shape) == 2:  # 
        kernel_padded = np.pad(kernel, [(0, image.shape[0] - kernel.shape[0]), 
                                        (0, image.shape[1] - kernel.shape[1])], 
                               mode='constant')

        kernel_fft = np.fft.fft2(kernel_padded)
        image_fft = np.fft.fft2(image)

        kernel_conj = np.conj(kernel_fft)
        wiener_result = (kernel_conj / (np.abs(kernel_fft)**2 + K)) * image_fft
        result = np.abs(np.fft.ifft2(wiener_result))
    else:  # 
        for i in range(image.shape[2]):  # 
            channel = image[:, :, i]
            kernel_padded = np.pad(kernel, [(0, channel.shape[0] - kernel.shape[0]), 
                                            (0, channel.shape[1] - kernel.shape[1])], 
                                   mode='constant')
            kernel_fft = np.fft.fft2(kernel_padded)
            channel_fft = np.fft.fft2(channel)
            kernel_conj = np.conj(kernel_fft)
            wiener_result = (kernel_conj / (np.abs(kernel_fft)**2 + K)) * channel_fft
            result[:, :, i] = np.abs(np.fft.ifft2(wiener_result))
    return result

# 
image_path = 'AUF6961.jpg'  # 
original_img = cv2.imread(image_path, cv2.IMREAD_COLOR)
original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)  #  BGR  RGB  Matplotlib 

# 
show_image("Original Image", original_img)

# 
kernel_size = 200  # 
angle = 90  # 
blurred_img, blur_kernel = motion_blur(original_img, kernel_size, angle)

# 
restored_img = wiener_filter(blurred_img, blur_kernel, K=0.01)

# 
show_image("Restored Image", restored_img)
```
